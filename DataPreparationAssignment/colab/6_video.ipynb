Distribution PLots

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(18, 6))

# Create a list of body parts for demonstration
body_parts = ["nose", "left_eye", "left_shoulder"]

# Plot distribution for x and y coordinates of selected body parts
for i, body_part in enumerate(body_parts, 1):
    plt.subplot(1, 3, i)
    sns.kdeplot(data[f"{body_part}_x"], shade=True, label=f"{body_part}_x")
    sns.kdeplot(data[f"{body_part}_y"], shade=True, label=f"{body_part}_y")
    plt.title(f"Distribution of {body_part} coordinates")
    plt.legend()

plt.tight_layout()
plt.show()
```

Here are the distribution plots for the x and y coordinates of the selected body parts:

Nose: The x and y coordinates have slightly different distributions. The y-coordinates seem to be more spread out than the x-coordinates.

Left Eye: Both x and y coordinates have relatively tight distributions, suggesting that the left eye's position doesn't vary as much.

Left Shoulder: The x and y distributions are distinct, with the y-coordinates having a wider spread than the x-coordinates.

```python
# Initialize the figure
plt.figure(figsize=(18, 6))

# Plot time series for x and y coordinates of selected body parts
for i, body_part in enumerate(body_parts, 1):
    plt.subplot(1, 3, i)
    plt.plot(data[f"{body_part}_x"], label=f"{body_part}_x", alpha=0.7)
    plt.plot(data[f"{body_part}_y"], label=f"{body_part}_y", alpha=0.7)
    plt.title(f"Time Series of {body_part} coordinates")
    plt.legend()

plt.tight_layout()
plt.show()
```

Here are the time series plots for the x and y coordinates of the selected body parts:

Nose: Both the x and y coordinates show some variability, suggesting movement or changes in position over time.

Left Eye: While the x-coordinate remains relatively stable, the y-coordinate exhibits more fluctuation, indicating vertical movement.

Left Shoulder: The x and y coordinates both fluctuate, indicating movement in both directions.

```python
# Initialize the figure
plt.figure(figsize=(18, 6))

# Plot scatter plots for x and y coordinates of selected body parts
for i, body_part in enumerate(body_parts, 1):
    plt.subplot(1, 3, i)
    sns.scatterplot(x=data[f"{body_part}_x"], y=data[f"{body_part}_y"], alpha=0.5, edgecolor=None)
    plt.title(f"Scatter Plot of {body_part} coordinates")

plt.tight_layout()
plt.show()
```

Here are the scatter plots for the x and y coordinates of the selected body parts:

Nose: There's a dense clustering in the middle, but some points are spread out, indicating variations in the nose's position.

Left Eye: The points form a relatively tight cluster, suggesting consistent positioning with minor deviations.

Left Shoulder: There's a wider spread of points, indicating more variability in the left shoulder's positioning.
