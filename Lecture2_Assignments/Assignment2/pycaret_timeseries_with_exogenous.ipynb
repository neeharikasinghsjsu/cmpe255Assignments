# **3. Get the dataset from github**

```python
import pandas as pd
dataset_url = 'https://raw.githubusercontent.com/neeharikasinghsjsu/cmpe255assignments/main/Lecture2_Assignments/Assignment2/dataset/macro_quarterly_ts_with_exogenous.csv'
dataset = pd.read_csv(dataset_url)
```

# **4. Scan the dataset**

```python
print("(Rows, Columns) in the dataset : ", dataset.shape, "\n\n")
dataset.head()
```

```python
dataset["date_index"] = pd.to_datetime(dataset['date'],format='%d/%m/%Y').dt.to_period('Q')
dataset
```

```python
dataset.drop(columns=["date"], inplace=True)
```

```python
dataset.set_index("date_index", inplace=True)
```

```python
dataset
```

# **5. Setup**

```python
from pycaret.time_series import *
setup(data=dataset, target="netexp", fh=3,numeric_imputation_target="ffill", numeric_imputation_exogenous="ffill",session_id=42, use_gpu=True)

```

# **6. Check Stats**

```python
check_stats()
```

# **6. Compare Models**

```python
compare_models()
```

```python
best = create_model('dt_cds_dt')
best
```

```python
plot_model(best)
```

```python
predict_model(best, fh=10)

```
