```python
import pandas as pd
dataset_url = 'https://raw.githubusercontent.com/neeharikasinghsjsu/cmpe255assignments/main/ApacheBeamAssignment/USA_cars_datasets.csv'
cars_data = pd.read_csv(dataset_url)
cars_data.head()
```

```python
cars_data.tail()
```

```python
cars_data.describe()
```

```python
missing_values = cars_data.isnull().sum()

missing_values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set_style("whitegrid")

# Plot the distribution of car prices
plt.figure(figsize=(12, 6))
sns.histplot(cars_data['price'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of car mileages
plt.figure(figsize=(12, 6))
sns.histplot(cars_data['mileage'], bins=50, kde=True, color='coral')
plt.title('Distribution of Car Mileages')
plt.xlabel('Mileage')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of the year of manufacture
plt.figure(figsize=(12, 6))
sns.histplot(cars_data['year'], bins=50, kde=True, color='mediumseagreen')
plt.title('Distribution of Year of Manufacture')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of car brands
plt.figure(figsize=(15, 8))
sns.countplot(y=cars_data['brand'], order=cars_data['brand'].value_counts().index, palette='viridis')
plt.title('Distribution of Car Brands')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.show()
```

```python
# Plot the distribution of states where the cars are located
plt.figure(figsize=(15, 8))
sns.countplot(y=cars_data['state'], order=cars_data['state'].value_counts().index, palette='magma')
plt.title('Distribution of Cars by State')
plt.xlabel('Count')
plt.ylabel('State')
plt.show()
```

```python
# Plot the distribution of car colors
plt.figure(figsize=(15, 8))
sns.countplot(y=cars_data['color'], order=cars_data['color'].value_counts().index, palette='rainbow')
plt.title('Distribution of Car Colors')
plt.xlabel('Count')
plt.ylabel('Color')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = cars_data[['price', 'year', 'mileage']].corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# 1. Distribution of car prices
# For a histogram, we'll just need the raw prices. However, for our purpose, we'll also get the frequency for each price range.

price_bins = [0, 5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 100000, 500000]
price_labels = ['0-5k', '5k-10k', '10k-15k', '15k-20k', '20k-25k', '25k-30k', '30k-35k', '35k-40k', '40k-45k', '45k-50k', '50k-100k', '100k-500k']
cars_data['price_range'] = pd.cut(cars_data['price'], bins=price_bins, labels=price_labels, right=False)

price_distribution = cars_data['price_range'].value_counts().sort_index().reset_index()
price_distribution.columns = ['price_range', 'count']

price_distribution
```

```python
# 2. Top car brands by frequency
brand_frequency = cars_data['brand'].value_counts().reset_index()
brand_frequency.columns = ['brand', 'count']

# We'll only take the top 10 brands for visualization
top_brands = brand_frequency.head(10)

top_brands
```

```python
# 3. Distribution of cars by year
year_distribution = cars_data['year'].value_counts().sort_index().reset_index()
year_distribution.columns = ['year', 'count']

year_distribution
```

```python
# 4. Mileage distribution of the cars
mileage_bins = [0, 10000, 20000, 30000, 40000, 50000, 100000, 150000, 200000, 300000, 500000, 1000000]
mileage_labels = ['0-10k', '10k-20k', '20k-30k', '30k-40k', '40k-50k', '50k-100k', '100k-150k', '150k-200k', '200k-300k', '300k-500k', '500k-1M']
cars_data['mileage_range'] = pd.cut(cars_data['mileage'], bins=mileage_bins, labels=mileage_labels, right=False)

mileage_distribution = cars_data['mileage_range'].value_counts().sort_index().reset_index()
mileage_distribution.columns = ['mileage_range', 'count']

mileage_distribution
```

```python
# 5. Car distribution across different states
state_distribution = cars_data['state'].value_counts().reset_index()
state_distribution.columns = ['state', 'count']

# Considering top states for a clearer visualization
top_states = state_distribution.head(10)

top_states
```

```python
price_data = price_distribution.to_json(orient='records')
print(price_data)

top_brands_data = top_brands.to_json(orient='records')
print(top_brands_data)

year_data = year_distribution.to_json(orient='records')
print(year_data)

mileage_data = mileage_distribution.to_json(orient='records')
print(mileage_data)

top_states_data = top_states.to_json(orient='records')
print(top_states_data)


```

```python
template = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>{title}</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
    <style>
        .bar {{
            fill: steelblue;
        }}
        .axis {{
            font-size: 10px;
        }}
        .chart-title {{
            font-size: 20px;
            font-weight: bold;
        }}
    </style>
</head>
<body>

<script>
    const data = {data};

    const margin = {{top: 20, right: 20, bottom: 50, left: 60}};
    const svgWidth = 700;
    const svgHeight = 500;
    const chartWidth = svgWidth - margin.left - margin.right;
    const chartHeight = svgHeight - margin.top - margin.bottom;

    const svg = d3.select('body')
        .append('svg')
        .attr('width', svgWidth)
        .attr('height', svgHeight);

    const chart = svg.append('g')
        .attr('transform', `translate(${{margin.left}}, ${{margin.top}})`);

    const yScale = d3.scaleLinear()
        .domain([0, d3.max(data, d => d.count)])
        .range([chartHeight, 0]);

    const xScale = d3.scaleBand()
        .domain(data.map(d => d.{key_name}))
        .range([0, chartWidth])
        .padding(0.3);

    const yAxis = d3.axisLeft(yScale);
    const xAxis = d3.axisBottom(xScale);

    chart.append('g')
        .attr('transform', `translate(0, ${{chartHeight}})`)
        .call(xAxis);

    chart.append('g')
        .call(yAxis);

    chart.append('text')
        .attr('class', 'chart-title')
        .attr('x', chartWidth / 2)
        .attr('y', 0)
        .attr('text-anchor', 'middle')
        .text("{title}");

    const bars = chart.selectAll('.bar')
        .data(data)
        .enter()
        .append('rect')
        .attr('class', 'bar')
        .attr('x', d => xScale(d.{key_name}))
        .attr('y', d => yScale(d.count))
        .attr('width', xScale.bandwidth())
        .attr('height', d => chartHeight - yScale(d.count))
        .attr('fill', (d, i) => `rgb(0, 0, ${{i * 25}})`);

    bars.append('title')
        .text(d => `Count: ${{d.count}}`);

</script>

</body>
</html>
"""
```
