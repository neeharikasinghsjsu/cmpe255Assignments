# **3. Get the dataset from github**

```python
import pandas as pd
dataset_url = 'https://raw.githubusercontent.com/neeharikasinghsjsu/cmpe255assignments/main/Lecture2_Assignments/Assignment2/dataset/market_dataset_association_rules.csv'
column_names = ["Date", "transaction_id", "item_description"]
dataset = pd.read_csv(dataset_url, names=column_names)
```

# **4. Scan the dataset**

```python
print("(Rows, Columns) in the dataset : ", dataset.shape, "\n\n")
dataset
```

# **5. Setup**

```python
from pycaret.arules import *
as_rules = get_rules(data = dataset, transaction_id = 'transaction_id', item_id = 'item_description')
```

```python
arules = create_model(metric='confidence', threshold=0.5, min_support=0.05)
```
