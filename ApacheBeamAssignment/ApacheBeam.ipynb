```python
import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions
```

```python
input_file = "apache_beam_vgsales.csv"

with beam.Pipeline(options=PipelineOptions()) as p:
    lines = p | "Read from CSV" >> beam.io.ReadFromText(input_file)
    lines | "Print lines" >> beam.Map(print)

```

```python
class ExtractYearAndSales(beam.DoFn):
    def process(self, element):
        try:
          _rank,_name,_platform,year,_genre,_publisher,_na_sales, _eu_sales,_jp_sales,_other_sales,global_sales= element.split(",")
          yield (int(year), float(global_sales))
        except:
          pass


class SkipFirstRow(beam.DoFn):
    def start_bundle(self):
        self._first_element = True

    def process(self, element):
        if self._first_element:
            self._first_element = False
        else:
            yield element

with beam.Pipeline(options=PipelineOptions()) as p:
    lines = p | "Read from CSV" >> beam.io.ReadFromText(input_file)
    # Apply the transform to skip the first row
    filtered_lines = lines | "Skip First Row" >> beam.ParDo(SkipFirstRow())
    sales_data = filtered_lines | "Extract Year and Sales" >> beam.ParDo(ExtractYearAndSales())
    sales_data | "Print data" >> beam.Map(print)
```

```python
class CalculateTotalSales(beam.PTransform):
    def expand(self, pcoll):
        return (
            pcoll
            | "Extract Year and Sales" >> beam.ParDo(ExtractYearAndSales())
            | "Group by Year" >> beam.GroupByKey()
            | "Sum Sales" >> beam.Map(lambda x: (x[0], sum(x[1])))
        )

with beam.Pipeline(options=PipelineOptions()) as p:
    lines = p | "Read from CSV" >> beam.io.ReadFromText(input_file)
    filtered_lines = lines | "Skip First Row" >> beam.ParDo(SkipFirstRow())
    totals = (filtered_lines
              | "Calculate Total Sales" >> CalculateTotalSales()
              | "Print Totals" >> beam.Map(print))
```

```python
with beam.Pipeline(options=PipelineOptions()) as p:
    lines = (
        p
        | "Read from CSV" >> beam.io.ReadFromText(input_file)
        | "Assign to Window" >> beam.WindowInto(
            beam.window.FixedWindows(60),
            trigger=beam.transforms.trigger.AfterWatermark(
                early=beam.transforms.trigger.AfterProcessingTime(5),
                late=beam.transforms.trigger.AfterCount(5)),
                accumulation_mode=beam.transforms.trigger.AccumulationMode.DISCARDING

            )
    )

    lines | "Print lines" >> beam.Map(print)
```

```python
output_file = "output.txt"

with beam.Pipeline(options=PipelineOptions()) as p:
    lines = p | "Read from CSV" >> beam.io.ReadFromText(input_file)
    totals = lines | "Calculate Total Sales" >> CalculateTotalSales()
    totals | "Write to Text" >> beam.io.WriteToText(output_file)
```
