# **3. Get the dataset from github**

```python
import pandas as pd
dataset_url = 'https://raw.githubusercontent.com/neeharikasinghsjsu/cmpe255assignments/main/Lecture2_Assignments/Assignment2/dataset/financials_anamoly.csv'
dataset = pd.read_csv(dataset_url)
```

# **4. Scan the dataset**

```python
print("(Rows, Columns) in the dataset : ", dataset.shape, "\n\n")
dataset.head()
```

# **4. Preprocess the data**

```python
dataset = dataset.replace({'\$': ''}, regex=True)
dataset = dataset.replace({'\,': ''}, regex=True)
dataset = dataset.replace({'\-': '0'}, regex=True)
dataset.head()
```

# **5. Divide the data**

```python
data = dataset.sample(frac=0.95, random_state=786)
data_unseen = dataset.drop(data.index).reset_index(drop=True)
data = data.reset_index(drop=True)

print("Data for modeling : ", data.shape)
print("Unseen data for predictions : ", data_unseen.shape)
```

```python
data.head()
```

```python
data_unseen.head()
```

# **6. Setup**

```python
from pycaret.anomaly import *
setup = setup(data, session_id = 123, use_gpu=True)
```

# **7. Create Model**

```python
models()
```

```python
financials_model = create_model('iforest')
financials_model
```

# **8. Assign Model**

```python
financials_anomalies = assign_model(financials_model)
financials_anomalies
```

# **8. Analyze Model**

```python
plot_model(financials_model, plot = 'tsne')
```

# **9. Predict Model**

```python
fa_prediction = predict_model(financials_model, data=data)
fa_prediction
```

# **10. Predict Model on unseen data**

```python
fa_prediction_unseen = predict_model(financials_model, data=data_unseen)
fa_prediction_unseen
```

# **10. Save Model**

```python
save_model(financials_model,'fa_model_V1')
```

# **11. Load Model for future predictions**

```python
loaded_fa_model = load_model('fa_model_V1')
loaded_fa_model
```
